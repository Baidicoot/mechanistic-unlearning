```python
# make pareto plots
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
import matplotlib.pyplot as plt
import pickle

# Assuming get_losses function is defined as in your snippet

def plot_pareto(x_axis_loss, y_axis_loss, x_range=None, y_range=None, include_text=False, plot_thresholds=True, plot_mask_ks=True, ):
    fig, ax = plt.subplots(figsize=(6, 6))  # Adjusted for better visibility of labels
    
    # Custom legend items
    custom_legend_items = [
        Line2D([0], [0], marker='^', color='w', label='Edge', markerfacecolor='k', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Weight', markerfacecolor='k', markersize=10),
        Line2D([0], [0], marker='s', color='w', label='Threshold', markerfacecolor='none', markeredgecolor='k', markersize=10),
        Line2D([0], [0], marker='s', color='w', label='Mask_k', markerfacecolor='k', markersize=10),
        Patch(facecolor='tab:blue', label='ACDCPP'),
        Patch(facecolor='tab:orange', label='CT'),
        Patch(facecolor='tab:green', label='None'),
    ]
    
    # Combine all paths for iteration
    all_paths = acdcpp_edge_paths + none_edge_paths + ct_edge_paths + acdcpp_weight_paths + none_weight_paths + ct_weight_paths
    all_dicts_reversed = {**acdcpp_edge_paths_dict_reversed, **ct_edge_paths_dict_reversed, **none_edge_paths_dict_reversed, **acdcpp_weight_paths_dict_reversed, **ct_weight_paths_dict_reversed, **none_weight_paths_dict_reversed}
    
    x_axis_losses = {}
    y_axis_losses = {}
    for path in all_paths:
        if 'acdcpp' in path:
            color = 'tab:blue'
        elif 'ct' in path:
            color = 'tab:orange'
        elif 'none' in path:
            color = 'tab:green'
        
        shape = '^' if 'edge' in path else 'o'  # square for edge models, circle for weight models
        fill_style = 'full' if "mask_k=" in path else 'none'
        losses = get_losses(path)
        label = all_dicts_reversed[path]

        if plot_thresholds and 'threshold=' in path:
            ax.scatter(losses[x_axis_loss], losses[y_axis_loss], color=color, marker=shape, facecolors=('none' if fill_style == 'none' else color))

            x_axis_losses[label] = losses[x_axis_loss]
            y_axis_losses[label] = losses[y_axis_loss]

            if include_text and (x_range is None or (losses[x_axis_loss] < x_range[1] and losses[x_axis_loss] > x_range[0])) and (y_range is None or (losses[y_axis_loss] < y_range[1] and losses[y_axis_loss] > y_range[0])):
                ax.text(losses[x_axis_loss], losses[y_axis_loss], ' ' + label, fontsize=7, ha='left', va='center', alpha=0.7)
        
        elif plot_mask_ks and "mask_k=" in path:
            ax.scatter(losses[x_axis_loss], losses[y_axis_loss], color=color, marker=shape, facecolors=('none' if fill_style == 'none' else color))

            x_axis_losses[label] = losses[x_axis_loss]
            y_axis_losses[label] = losses[y_axis_loss]

            if include_text and (x_range is None or (losses[x_axis_loss] < x_range[1] and losses[x_axis_loss] > x_range[0])) and (y_range is None or (losses[y_axis_loss] < y_range[1] and losses[y_axis_loss] > y_range[0])):
                ax.text(losses[x_axis_loss], losses[y_axis_loss], ' ' + label, fontsize=7, ha='left', va='center', alpha=0.7)
    
    # Adjusting x and y axis limits if ranges are provided
    if x_range is not None:
        ax.set_xlim(x_range)
    if y_range is not None:
        ax.set_ylim(y_range)
    
    # Setting labels, title
    ax.set_xlabel(x_axis_loss)
    ax.set_ylabel(y_axis_loss)
    ax.set_title(f'Pareto Plot of Models of {x_axis_loss} vs {y_axis_loss}')
    
    # Adding custom legend for color, shape, and fill style
    ax.legend(handles=custom_legend_items, loc='best', title="Legend")
    
    plt.show()
    return x_axis_losses, y_axis_losses
```
