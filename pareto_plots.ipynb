```python
import pickle
def get_losses(path):
    with open(f"{path}/ckpts/final_losses.pkl", "rb") as f:
        _, test_losses = pickle.load(f)
    test_losses_dict = {}
    for loss_type, loss_list in test_losses.items():
        try:
            test_losses_dict[loss_type] = loss_list[-1].item()
        except AttributeError:
            test_losses_dict[loss_type] = loss_list[-1]
    return test_losses_dict

get_losses(none_edge_paths[0])
```

```python
# make pareto plots
x_axis_loss = "ioi_logit_diff"
y_axis_loss = "owt"

import matplotlib.pyplot as plt
import pickle

# Assuming get_losses function is defined as in your snippet

def plot_pareto():
    fig, ax = plt.subplots()
    
    # Plotting edge models
    for path in acdcpp_edge_paths + none_edge_paths:
        color = 'orange' if 'acdcpp' in path else 'blue'
        shape = 's'  # square for edge models
        losses = get_losses(path)
        ax.scatter(losses[x_axis_loss], losses[y_axis_loss], color=color, marker=shape, label='ACDCPP Edge' if 'acdcpp' in path else 'None Edge')
    
    # Plotting weight models
    for path in acdcpp_weight_paths + none_weight_paths:
        color = 'orange' if 'acdcpp' in path else 'blue'
        shape = 'o'  # circle for weight models
        losses = get_losses(path)
        ax.scatter(losses[x_axis_loss], losses[y_axis_loss], color=color, marker=shape, label='ACDCPP Weight' if 'acdcpp' in path else 'None Weight')
    
    # Setting labels, title, and legend
    ax.set_xlabel(x_axis_loss)
    ax.set_ylabel(y_axis_loss)
    ax.set_title('Pareto Plot of Models')
    handles, labels = ax.get_legend_handles_labels()
    unique = [(h, l) for i, (h, l) in enumerate(zip(handles, labels)) if l not in labels[:i]]
    ax.legend(*zip(*unique))
    
    plt.show()

plot_pareto()
```
