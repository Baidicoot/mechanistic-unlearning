```python
import pickle
def get_losses(path):
    with open(f"{path}/ckpts/final_losses.pkl", "rb") as f:
        _, test_losses = pickle.load(f)
    test_losses_dict = {}
    for loss_type, loss_list in test_losses.items():
        try:
            test_losses_dict[loss_type] = loss_list[-1].item()
        except AttributeError:
            test_losses_dict[loss_type] = loss_list[-1]
    return test_losses_dict

get_losses(none_edge_paths[0])
```

```python
# make pareto plots
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
import matplotlib.pyplot as plt
import pickle

# Assuming get_losses function is defined as in your snippet

def plot_pareto(x_axis_loss, y_axis_loss, x_range=None, y_range=None):
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Custom legend items
    custom_legend_items = [
        Line2D([0], [0], marker='s', color='w', label='Edge', markerfacecolor='k', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Weight', markerfacecolor='k', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Threshold', markerfacecolor='none', markeredgecolor='k', markersize=10),
        Line2D([0], [0], marker='o', color='w', label='Mask_k', markerfacecolor='k', markersize=10),
        Patch(facecolor='tab:blue', label='ACDCPP'),
        Patch(facecolor='tab:orange', label='CT'),
        Patch(facecolor='tab:green', label='None'),
    ]
    
    x_losses_dict = {}
    y_losses_dict = {}
    # Plotting edge models
    for path in acdcpp_edge_paths + ct_edge_paths + none_edge_paths:
        if 'acdcpp' in path:
            color = 'tab:blue'
        elif 'ct' in path:
            color = 'tab:orange'
        elif 'none' in path:
            color = 'tab:green'
        
        shape = 's'  # square for edge models
        fill_style = 'full' if "mask_k=" in path else 'none'
        losses = get_losses(path)
        ax.scatter(losses[x_axis_loss], losses[y_axis_loss], color=color, marker=shape, facecolors=('none' if fill_style == 'none' else color))

        x_losses_dict[path] = losses[x_axis_loss]
        y_losses_dict[path] = losses[y_axis_loss]
    
    # Plotting weight models
    for path in acdcpp_weight_paths + ct_weight_paths + none_weight_paths:
        if 'acdcpp' in path:
            color = 'tab:blue'
        elif 'ct' in path:
            color = 'tab:orange'
        elif 'none' in path:
            color = 'tab:green'
        
        shape = 'o'  # circle for weight models
        fill_style = 'full' if "mask_k=" in path else 'none'
        losses = get_losses(path)
        ax.scatter(losses[x_axis_loss], losses[y_axis_loss], color=color, marker=shape, facecolors=('none' if fill_style == 'none' else color))

        x_losses_dict[path] = losses[x_axis_loss]
        y_losses_dict[path] = losses[y_axis_loss]
    
    # Adjusting x and y axis limits if ranges are provided
    if x_range is not None:
        ax.set_xlim(x_range)
    if y_range is not None:
        ax.set_ylim(y_range)
    
    # Setting labels, title
    ax.set_xlabel(x_axis_loss)
    ax.set_ylabel(y_axis_loss)
    ax.set_title(f'Pareto Plot of Models of {x_axis_loss} vs {y_axis_loss}')
    
    # Adding custom legend
    ax.legend(handles=custom_legend_items, loc='best')
    print(x_losses_dict)
    print(y_losses_dict)
    plt.show()
    return x_losses_dict, y_losses_dict

plot_pareto(x_axis_loss = "ioi_logit_diff", y_axis_loss = "owt")
```
